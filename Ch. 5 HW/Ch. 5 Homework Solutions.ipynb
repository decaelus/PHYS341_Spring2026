$U(r) = A \bigg[\big(e^{(R-r)/S}-1\big)^2-1\bigg]$

Here's a plot of this potential:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#v = (R-r)/S
v = np.linspace(-1, 5, 1000)
A = 1.0
U = A*((np.exp(-v)-1)**2-1)
plt.plot(v, U, lw=3)
plt.xlabel("$(R-r)/s$", fontsize=18)
plt.ylabel("$U(r)$", fontsize=18)
plt.axhline(0, color='red', ls='--', lw=3)
plt.title("Potential from Problem 5.2", fontsize=18)
```

Equilibrium position $r_\mathrm{0}$ at minimum of $U(r)$ occurs where $\frac{dU}{dr} = 0$.

$\frac{dU}{dr} = 2A \big(e^{(R - r_\mathrm{0})/S}-1\big)e^{(R - r_\mathrm{0})/S}\left(\frac{-1}{S}\right)=0, $ which can only be satisifed when $\big(e^{(R - r_\mathrm{0})/S}-1\big) = 0 \Rightarrow r_\mathrm{0} = R$. 

The plot shows this is a minimum, so no need to check $\frac{d^2U}{dr^2}$.

For $r = r_\mathrm{0} + x = R + x$ with $x$ small, $ U(r = R + x) \approx U(r = R) + \left(\frac{dU}{dr}\right)_{r = R} x + \frac{1}{2} \left(\frac{d^2 U}{dr^2}\right)_{r = R} x^2. $  

$\left(\frac{dU}{dr}\right)_{r = R} = 0$ since $r = R$ is a minimum.

To simplify calculation of $\left(\frac{d^2 U}{dr^2}\right)$, define $-v \equiv -\left(\frac{r - R}{S}\right) \Rightarrow dv = \frac{dr}{S}$. 

$ \frac{dU}{dv} = -2 A \left(e^{-v} - 1\right)e^{-v}\\ \frac{d^2U}{dv^2}_{v = 0} = +2 A \left[e^{-2v} + \left(e^{-v} -1\right)e^{-v}\right]_{v=0} = 2A, $ 

which, by the way, is greater than 0, confirming $r=R$ is a minimum for $U(r)$. $\Rightarrow \boxed{U(r = R + x) \approx \text{const.} + \frac{1}{2} \left(2 A /S^2\right)x^2}$

#### 5.11 

The total energy $E$ of a simple harmonic oscillator is $ E = \frac{1}{2} m \dot{x}^2 + \frac{1}{2} m \omega^2 x^2, $ where $m$ is the mass and $\omega$ the natural frequency. If we know the oscillating mass has velocity $v_\mathrm{1}/v_\mathrm{2}$ at $x_\mathrm{1}/x_\mathrm{2}$, we know $E$ at two points, which is enough to solve for $\omega$:

$ \frac{1}{2} m v_\mathrm{1}^2 + \frac{1}{2} m \omega^2 x_\mathrm{1}^2 = \frac{1}{2} m v_\mathrm{2}^2 + \frac{1}{2} m \omega^2 x_\mathrm{2}^2\\ \Rightarrow v_\mathrm{1}^2 - v_\mathrm{2}^2 = \left(x_\mathrm{2}^2 - x_\mathrm{1}^2\right)\omega^2 $

$\Rightarrow \boxed{\omega = \sqrt{\frac{v_\mathrm{1}^2 - v_\mathrm{2}^2}{x_\mathrm{2}^2 - x_\mathrm{1}^2}}}$.

And the amplitude $A$ is related to $E$ as $E = \frac{1}{2} m \omega^2 A^2$, so $\boxed{A = \sqrt{\left(\frac{x_\mathrm{2}^2 - x_\mathrm{1}^2}{v_\mathrm{1}^2 - v_\mathrm{2}^2}\right)v_\mathrm{1}^2 + x_\mathrm{1}^2}}$

(b) The energy dissipated by the damping $\Delta E$ is the work done over one cycle: <br><br>  $\Delta E = \int_P F_\mathrm{damp} \upsilon dt$, the integral over one period $P$. <br><br>  $\Delta E = \int_P 2 m \beta \upsilon\ \upsilon dt = 2 m \beta \int_P \upsilon^2 dt = 2 m \beta A^2 \omega^2 \int_P \sin^2\left(\omega t - \delta\right) dt$. <br><br>  Using Euler's identity, $\sin^2\left(\phi\right) = \left[ \frac{1}{2i}\left( e^{i \phi} - e^{-i \phi} \right) \right]^2 = -\frac{1}{4}\left[ e^{2i \phi} + e^{-2i \phi} - 2 \right] = \frac{1}{2} \left[1 - \cos\left( 2 \phi \right) \right]$, and then we can do the integral. <br><br>  With the substitution $\phi = \omega t - \delta$ (and $dt = d\phi/\omega$), the integral becomes <br><br>  $\Delta E = m \beta A^2 \omega \int_{\phi = -\delta}^{2\pi - \delta} \left[ 1 - \cos\left( 2 \phi \right) \right] d\phi = m \beta A^2 \omega \bigg[ \phi -\frac{1}{2}\sin\left( 2 \phi \right) \bigg]_{-\delta}^{2\pi - \delta} = \boxed{ 2 \pi m \beta A^2 \omega }$.

(c) $2\pi E/\Delta E = 2\pi \frac{1}{2} m \omega^2 A^2/\left( 2\pi m \omega \beta A^2 \right) = \boxed{ \omega / 2\beta = Q }$.

#### 5.50

The function shown in Figure 5.28(b) is odd, meaning all the cosine terms are zero or $a_\mathrm{n} = 0$. 

To calculate $b_\mathrm{n}$, I need to calculate the following integral: $\frac{2}{\tau}\int_{-\tau/2}^{\tau/2} f(t) \sin(n \omega t) dt$.  Between -1 and 1, $f(t) = t$. The period of $f(t)$ also sets the period for the Fourier series, i.e. $\tau = 2$ and $\omega = 2\pi/2 = \pi$.<br><br>  

We can calculate the anti-derivative using integration by parts. Remember $\int{d(u\ v)} = \int{du\ v} + \int{u\ dv} \Rightarrow \int{du\ v} = \int{d(u\ v)} - \int{u\ dv} = \left[u\ v\right]_\mathrm{limits} - \int{u\ dv}$. 

$\int_{-1}^{+1} t \sin(\pi n t) dt = \left[ -t \cos(\pi n t)/(\pi n) \right]_{-1}^{+1} + \int_{-1}^{+1} \cos(\pi n t)/(n \omega) dt = \left[ -t \cos(\pi n t)/(\pi n) + \sin(\pi n t)/(\pi n)^2 \right]_{-1}^{+1} = -\frac{2 \cos(\pi n)}{\pi n}$.

```python
%matplotlib inline

import numpy as np
import pylab as pl

def sawtooth(x, fmax = 1., period = 2.):
    x_mod = (x + 0.5*period) % period
    return fmax*x_mod - fmax

def fourier_series(x, num_terms = 2, omega = np.pi):
    fs = np.zeros_like(x)
    for n in range(1, num_terms + 1):
        bn = -2./np.pi/n*np.cos(np.pi*n)
        fs += bn*np.sin(n*omega*x)
    return fs

fig = pl.figure(figsize=(8,8))
ax = fig.add_subplot(111)

x = np.linspace(-1.5, 3, 100)
ax.plot(x, sawtooth(x), lw=3, label="sawtooth")
ax.axhline(0., color='black', lw=3)
ax.axvline(0., color='black', lw=3)
ax.tick_params(axis='both', labelsize=24)

ax.plot(x, fourier_series(x, num_terms = 2), ls='--', lw=3, label='2 terms')
ax.plot(x, fourier_series(x, num_terms = 10), lw=3, ls='-.', label='10 terms')

ax.legend(loc='best' , fontsize=24)
```

#### 5.53

(a) First step is to find the Fourier series representing the driving force. The force is an even function, so all the sine terms are zero, and we need only calculate the $a_\mathrm{n}$:

$a_\mathrm{n} = \frac{2}{\tau}\int_{-\tau/2}^{\tau/2} f(t) \cos(n \omega t) dt = \int_{-1}^{0} \left( 1 + t \right) \cos(n \pi t) dt - \int_{0}^{+1}\left( 1 - t \right) \cos(n \pi t) dt = \frac{2 ( 1 - \cos(n \pi) )}{(n\pi)^2}$.

As described in Section 5.8, the solution $x(t) = \sum_{n = 0}^\infty A_\mathrm{n} \cos(n \omega t - \delta_\mathrm{n})$, with $A_\mathrm{n} = \frac{f_\mathrm{n}}{\sqrt{(\omega_0^2 - n^2\omega^2)^2 + 4 \beta^2 n^2 \omega^2}}$ and $\delta_\mathrm{n} = \arctan\bigg(\frac{2 \beta n \omega}{\omega_0^2 - n^2 \omega^2}\bigg)$. 

The plot for the solution is below.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl

def fn(n):
    return 2.*(1 - np.cos(n*np.pi))/(n*np.pi)**2

def fourier_series(x, num_terms = 2, omega = np.pi):
    fs = np.ones_like(x)*0.5
    for n in range(1, num_terms + 1):
        an = fn(n)
        fs += an*np.cos(n*omega*x)
    return fs

def An(n, omega0, omega, beta):
    return fn(n)/np.sqrt((omega0**2 - (n*omega)**2)**2 + 4*(beta*n*omega)**2)

def deltan(n, omega0, omega, beta):
    ret_val = np.pi/2.
    if((omega0**2 - (n*omega)**2) != 0.):
        ret_val = np.arctan(2.*beta*n*omega/(omega0**2 - (n*omega)**2))
    
    return ret_val

def soln(t, omega0, omega, beta, num_terms = 4):
    x = np.ones_like(t)*0.5
    
    for n in range(1, num_terms + 1):        
        x += An(n, omega0, omega, beta)*np.cos(omega*t - deltan(n, omega0, omega, beta))
        
    return x

def driving_force(t, tau, fmax = 1.):
    t_mod = (t % tau) - tau
   
    ret_val = np.zeros_like(t_mod)
    ret_val[(t_mod > -1.0)] = fmax*t_mod[(t_mod > -1.0)] + 1.0
    ret_val[(t_mod < -1.0)] = -fmax*(t_mod[(t_mod < -1.0)] + 2.) + 1.0

    return ret_val

tau0 = 2.
omega0 = 2.*np.pi/tau0
beta = 0.1
#driving period
tau = 2.
omega = 2.*np.pi/tau

fig = pl.figure(figsize=(8,8))
ax = fig.add_subplot(111)

t = np.linspace(-1, 3, 100)
ax.plot(t, driving_force(t, tau), label='driving force', lw=3)
print("solution -- ")
ax.plot(t, fourier_series(t, num_terms=4), label='driving force (Fourier series)', 
        lw=6, ls='--', alpha=0.5)
x = soln(t, omega0, omega, beta)
ax.plot(t, x, lw = 3, ls='-.', label='oscillator')
ax.set_xlabel('time', fontsize=18)
ax.legend(loc='best', fontsize=18)
ax.tick_params(axis='both', labelsize=16)
```
